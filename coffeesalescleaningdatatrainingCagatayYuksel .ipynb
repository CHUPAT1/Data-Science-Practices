# 1.Kütüphaneleri İçeri Aktarma

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2.Verileri İçeri Aktarma

```python
df = pd.read_csv('/kaggle/input/cafe-sales-dirty-data-for-cleaning-training/dirty_cafe_sales.csv')
```

# 3.Verileri İnceleme

```python
df = df.dropna()
df.isna()
```

* ****Na olan değerleri verimizden çıkardık****

```python
df.head(200)
```

* ****Hala bilinmeyen verilerimiz var****

# 4.Veriyi Temizleme

```python
df_cleaned = df[df['Payment Method'] != 'UNKNOWN']
df_cleaned = df_cleaned[df_cleaned['Payment Method'] != 'ERROR']
df_cleaned = df_cleaned[df_cleaned['Location'] != 'UNKNOWN']
df_cleaned = df_cleaned[df_cleaned['Total Spent'] != 'ERROR']
df_cleaned = df_cleaned[df_cleaned['Total Spent'] != 'UNKNOWN']
df_cleaned = df_cleaned[df_cleaned['Quantity'] != 'ERROR']
df_cleaned = df_cleaned[df_cleaned['Quantity'] != 'UNKNOWN']
df_cleaned = df_cleaned[df_cleaned['Item'] != 'UNKNOWN']
df_cleaned = df_cleaned[df_cleaned['Item'] != 'ERROR']
df_cleaned = df_cleaned[df_cleaned['Price Per Unit'] != 'UNKNOWN']
df_cleaned = df_cleaned[df_cleaned['Price Per Unit'] != 'ERROR']
```

```python
df_cleaned = df_cleaned.sort_values(by = 'Price Per Unit')
df_cleaned.head(200)
```

* ****Veriyi temizledik****

# 5.Veriyi Analiz Etme
* ****İlk başta ürünlerin ve fiyatlarını karşılaştıralım****

```python
plt.figure(figsize=(8, 5))
plt.bar(df_cleaned['Item'], df_cleaned['Price Per Unit'])
plt.title = 'Ürünler ve Fiyatları'
plt.xlabel = 'Ürün'
plt.ylabel = 'Fiyat'
plt.show()
```
